```python
w = 0.5
input = 0.5
y = 0.8

h = 0.001

for i in range(1101):
    pred = input * w
    error = (pred - y) ** 2

    print(f"{i}: Error: {error} - Prediction: {pred}")

    # Try go up and go down
    up_pred = input * (w + h)
    up_error = (y - up_pred) ** 2

    down_pred = input * (w - h)
    down_error = (y - down_pred) ** 2

    # check which is better
    if(down_error < up_error):
        w = w - h
    else:
        w = w + h

```
